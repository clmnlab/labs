```python
SRplot_L=SRplotdat[SRplotdat.hand=='L']
SRplot_R=SRplotdat[SRplotdat.hand=='R']
SRplot_day1Train=SRplotdat[SRplotdat.TEST=='Day1-Training']
SRplot_day1Test=SRplotdat[SRplotdat.TEST=='Day1-Test']
SRplot_day2Train=SRplotdat[SRplotdat.TEST=='Day2-Training']
SRplot_day2PreTest=SRplotdat[SRplotdat.TEST=='Day2-PreTest']
SRplot_day2PostTest=SRplotdat[SRplotdat.TEST=='Day2-PostTest']
SRplot_mean=SRplotdat.groupby('mean_trial').mean()

fig=px.line(SRplotdat,x='trial',y='SuccessRate',color='hand',line_group='TEST')
fig.add_shape(type="rect",x0=80,y0=0,x1=84,y1=1.0,layer='below',line=dict(color="black",dash='dot',width = 2),name='Day1-Test')
fig.add_shape(type="rect",x0=84,y0=0,x1=88,y1=1.0,layer='below',line=dict(color="black",dash='dot',width = 2),name='Day2-PreTest')
fig.add_shape(type="rect",x0=168,y0=0,x1=180,y1=1.0,layer='below',line=dict(color="black",dash='dot',width = 2),name='Day2-PostTest')
for i in range(8,80,8):
    fig.add_shape(dict(type='line',x0=i,y0=0,x1=i,y1=1.0,line=dict(color='green',dash='dashdot',width = 1),name='BLOCK'))
for i in range(96,168,8):
    fig.add_shape(dict(type='line',x0=i,y0=0,x1=i,y1=1.0,line=dict(color='green',dash='dashdot',width = 1),name='BLOCK'))


fig.add_trace(go.Scatter(x=[80,86,170],y=[1.05,1.1,1.1],text=["DAY1-Test","DAY2-PreTest","DAY2-PostTest"],mode="text",name='TEST'))
fig.update_layout(
    title={'text': "SuccessRate",'y':0.95,'x':0.5,'xanchor': 'center','yanchor': 'top'})
fig.update_xaxes(tickvals=list(range(8,81,8))+[84,88]+list(range(96,169,8))+[180])
fig.show()
fig.write_html("GAB-pilot_GAB-A-98-SuccessRate.html")
```
